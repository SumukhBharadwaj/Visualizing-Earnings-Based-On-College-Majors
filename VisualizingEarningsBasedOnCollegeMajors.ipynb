# Visualizing Earnings Based On College Majors

In this project, we will explore the dataset on job outcomes of students who graduated from college between 2010 and 2012 using data visualization.

FiveThirtyEight, a website that focuses on opinion poll analysis, politics and economics cleaned the original dataset released by American Community Survey and released it on their Github repo

The following is the link to the dataset on FiveThirtyEight's Github repo: [College Majors](https://github.com/fivethirtyeight/data/tree/master/college-majors)

The goal of this project is to practice creating different types of plots using matplotlib and pandas.


## Reading the Dataset

```python
# Reading the file into a dataframe #
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_graduates = pd.read_csv("recent-grads.csv")
print(recent_graduates.iloc[0])
print(recent_graduates.head(4))
print(recent_graduates.tail(4))
recent_graduates.describe()
```

```python
# Count the number of rows in the datafrome #
no_of_rows_before_drop = recent_graduates.shape
print(no_of_rows_before_drop)
```

```python
# Drop the rows with missing values #
recent_graduates = recent_graduates.dropna()
recent_graduates
```

## Scatter Plots To Explore Relationships Between Columns



```python
# Sample_size column vs Median column #
ax1 = recent_graduates.plot(x="Sample_size", y="Median", kind = "scatter")
ax1.set_title("Sample_size vs Median")
```

```python
# Sample_size vs Unemployment_rate #
ax2 = recent_graduates.plot(x= "Sample_size", y= "Unemployment_rate", kind = "scatter")
ax2.set_title("Sample_size vs Unemployment_rate")
```

```python
# Full_time vs Median #
ax3 = recent_graduates.plot(x= "Full_time", y="Median", kind= "scatter")
ax3.set_title("Full_time vs Median")
```

```python
# ShareWomen vs Unemployment_rate #
ax3 = recent_graduates.plot(x= "ShareWomen", y="Unemployment_rate", kind= "scatter")
ax3.set_title("ShareWomen vs Unemployment_rate")
```

```python
# Men vs Median #
ax3 = recent_graduates.plot(x= "Men", y="Median", kind= "scatter")
ax3.set_title("Men vs Median")
```

```python
# Women vs Median #
ax3 = recent_graduates.plot(x= "Women", y="Median", kind= "scatter")
ax3.set_title("Women vs Median")
```

## Histograms To Explore Distributions Of Columns


```python
# Sample_size Column #
recent_graduates["Sample_size"].plot(kind = "hist", bins = 20, range= (0,5000))
```

```python
# Median Column #
recent_graduates["Median"].plot(kind = "hist", bins = 20, range = (0,110000))
```

```python
# Employed Column #
recent_graduates["Employed"].plot(kind = "hist", bins = 20, range= (0,400000))
```

```python
# Full_time Column #
recent_graduates["Full_time"].plot(kind = "hist", bins = 20, range= (0,300000))
```

```python
# ShareWomen Column #
recent_graduates["ShareWomen"].plot(kind = "hist", bins = 10, range= (0.0,1.0))
```

```python
# Unemployment_rate #
recent_graduates["Unemployment_rate"].plot(kind = "hist", bins = 10, range= (0.00, 0.18))
```

```python
# Men #
recent_graduates["Men"].plot(kind = "hist", bins = 10, range= (0.0,200000))
```

```python
# Women #
recent_graduates["Women"].plot(kind = "hist", bins = 10, range= (0,350000))
```

## Scatter Matrix Plots to Visualize Relationships And Distributions

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_graduates[["Sample_size", "Median"]], figsize = (10,10))
```

```python
scatter_matrix(recent_graduates[["Sample_size", "Median", "Unemployment_rate"]], figsize = (10,10))
```
